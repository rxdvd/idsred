````python
import os
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

from idsred import set_workdir, collect_data
from idsred.spec2Dreduc import quick_2Dreduction
from idsred.spec1Dreduc import quick_1Dreduction
from idsred.wavecalib import find_wavesol, apply_wavesol
from idsred.fluxcalib import fit_sensfunc, calibrate_spectra
from idsred.utils import plot_image, fits2array
````

## Setting the working directory

The first step is to tell the pipeline where to find all the images of a given night (e.g. 2022-10-08). This is important as these are constantly used in the processes below.
In addition, a ``processing`` directory is created under the working directory where all the outputs of the reduction are saved.

````python
set_workdir('20221008')
````

If you which, you can take a look at the files.

````python
observations = collect_data()
observations.summary
````

## 2D reduction

The next step is the 2D reduction of the images: bias subtraction, flat fielding, combining images of the same object (average combine by default) and cosmic rays rejection (for the science images).

The pipeline automatically identifies each type of image by looking at the headers and creates master images for the BIAS, FLAT and ARC, where the latter is used for the wavelength solution (see bellow). A quick reduction method can be done with ``quick_2Dreduction``, which should usually work correctly:

````python
quick_2Dreduction()
````

Overscan subtraction and image trimming are done by default (recommended), but these parameters can be changed.

The reduced images are saved under the ``processing`` directory with the ``_2d.fits`` extension. **Note** that the original files (e.g. ``r1634461.fit``) have the data in the first *header data unit* (HDU) extension (index 1), but the output files have the data in the *primary header unit* (PHU; index 0).

To check one of the output files, simply use ``plot_image``:

````python
hdu = fits.open('20221008/processing/SP0305+261_2d.fits')
plot_image(hdu[0].data)
````

For a manual 2D reduction, you can use the following functions:

```
create_master_arc(observations, beginning=True, method=method)
master_bias = create_master_bias(observations, subtract_overscan, trim_image, method)
master_flat = create_master_flat(observations, master_bias, subtract_overscan, trim_image, method)
red_images = reduce_images(observations, master_bias, master_flat, subtract_overscan, trim_image, method)
```

## 1D redcution

Now comes the 1D reduction, i.e. extracting the trace where the spectra are found. For a quick reduction use ``quick_1Dreduction`` (check the parameters for diagnostic plots):

````python
quick_1Dreduction(plot_trace=True)
````

The reduced images are saved under the ``processing`` directory with the ``_1d.fits`` extension. At this point, we should have (uncalibrated) spectra with familiar shapes.

For a manual 1D reduction, you can follow this example:

```
image = ccdproc.CCDData.read(fits_file) 

# extract trace
raw_spectrum = optimised_trace(image, plot_diag=False, plot_trace=True)  # or quick_trace()
image.data = raw_spectrum

object_name = os.path.basename(fits_file).split('_')[0]
outfile = f'{object_name}_1d.fits'
image.write(outfile, overwrite=True)
```

## Wavelength calibration

Here we find the *wavelength solution* to calibrate the wavelength of the science images. This is easily done with ``find_wavesol``, where you can specify the function used (*legendre* or *chebyshev*), order of the polynomial and sigma clipping (initial coefficient values are fixed). Usually, a dispersion of less than 1 Angstrom is wanted. This function uses the master ARC file created during the 2D reduction, where a set ARC files, with different exposure times, from the beginning of the night are combined. This is an important step as it allows the pipeline to identify spectral lines along the whole wavelength range.

````python
find_wavesol(func='legendre', sigclip=3, k=3, plot_solution=True)
````

In this example, we see that the dispersion is 0.2 Angstrom. The blue crosses in the bottom plot are lines rejected via sigma clipping. 

The wavelength solution is saved under the ``processing`` directory (``wavesol.txt``). Note that if one night is missing the wavelength solution, this can be copied from a different night, but it is not recommended unless there is no other option.

## Flux calibration

The last step is to get the *sensitivity function* for the flux calibration. This is created by comparing the spectrum of an observed standrd, with the calibrated/catalog version of it. The INT standard are usually low resolution and not found in other catalogs (not good!), but they still work. The sensitivity function can be obtained with the ``fit_sensfunc`` function, where a polynomial is fit:

````python
fit_sensfunc(degree=8, wmin=3800, wmax=9000, plot_diag=True, use_master_arc=False)
````

The ``xmin`` and ``xmax`` parameters trim the edges of the spectra, which allows you to avoid ranges where there is no signal (i.e. just background noise) and helps in estimating the sensitivity function. The result is saved under the ``processing`` directory (``sensfunc.txt``).

Now that the flux calibration is done, we can calibrate all the science images with the function ``calibrate_spectra``, where the outputs have the extension ``_wf.fits``:

````python
calibrate_spectra(use_master_arc=False)
````

Let's see the results:

````python
hdu = fits.open('20221008/processing/2022VQZ_wf.fits')
wave, flux = fits2array(hdu)

plt.plot(wave, flux)
plt.show()
````

## Summary

For a quick reduction, assuming everything works smoothly, simply use the following functions:
    

```
set_workdir()
quick_2Dreduction()
quick_1Dreduction()
find_wavesol()
fit_sensfunc()
calibrate_spectra()
```

The default parameters might not be enough (e.g. when fitting a polynomial), so just play around with them while using the diagnostic plots.
