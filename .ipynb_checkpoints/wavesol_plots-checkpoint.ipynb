```python
import os
import numpy as np

import astropy.units as u
from astropy.io import fits

import matplotlib.pyplot as plt
```

Wavelength identification taken from https://www.ing.iac.es/astronomy/observing/manuals/ps/tech_notes/tn070.pdf, specially Tables 4 & 5

```python
hdu = fits.open('static/master_arc.fits')
arc_data = hdu[0].data
profile = np.average(arc_data[:, 150:190], axis=1)[::-1]

cut = 1950
xmin = 1000
xmax = 3750

blue_profile = profile[xmin:cut]
blue_columns = np.arange(xmin, cut)
red_profile = profile[cut:xmax]
red_columns = np.arange(cut, xmax)
```

```python
start_adding = False
with open('static/idmaster_arc') as arcfile:
    for line in arcfile.readlines():
        if 'function' in line:
            break  # finish
        if start_adding is True:
            iraf_cols.append(float(line.split()[0]))
            iraf_waves.append(float(line.split()[1]))
        if 'features' in line:
            iraf_cols, iraf_waves = [], []
            start_adding = True
```

```python
def plot_arc(columns, profile, iraf_cols, iraf_waves, red=False):
    
    fig, ax = plt.subplots(figsize=(14, 6))
    plt.plot(columns, profile, c='k')

    # plot lines
    for i, (col, wave) in enumerate(zip(iraf_cols, iraf_waves)):
        diff = np.abs(columns-col)
        if diff.min() < 3:
            ind = np.argmin(diff)
            string = f'{wave:.1f} (col. {col:.0f})'
            if red is False:
                plt.scatter(col-2, profile[ind]*1.05, marker='|', s=300, zorder=10, color='r')
                plt.text(col-5, profile[ind]*1.12, string, rotation='vertical', fontsize=16)
            else:
                if i%2==0:
                    plt.scatter(col-2, profile[ind]*1.1, marker='|', s=300, zorder=10, color='r')
                    plt.text(col-5, profile[ind]*1.2, string, rotation='vertical', fontsize=16)
```

```python
plot_arc(red_columns, red_profile, iraf_cols, iraf_waves, red=True)
plt.tight_layout()
plt.savefig('static/arc_wavelengths_red.jpg')
```

```python
plot_arc(blue_columns, blue_profile, iraf_cols, iraf_waves)
plt.tight_layout()
plt.savefig('arc_wavelengths_blue.jpg')
```
